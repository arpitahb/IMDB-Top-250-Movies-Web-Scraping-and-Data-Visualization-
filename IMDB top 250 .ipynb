```python
year_tags= soup.find_all('span', {'class':"secondaryInfo"})
```

```python
 rank_tags= soup.find_all('td', {'class':"titleColumn"})
```

```python
rank=[]
for r in rank_tags:
    rank.append(r.get_text(strip=True).split('.')[0])
```

```python
rank
```

```python
ratings_tags=soup.find_all('td',{'class':"ratingColumn imdbRating"})
```

```python
rating=[]
for R in ratings_tags:
    rating.append(R.strong.text)
```
